```python
import pandas as pd
import numpy as np
from plotnine import *

df = pd.read_csv('Distribution_Data.csv')
density_plot=(ggplot(df,aes(x="value",fill="class"))
    +geom_density(alpha=1)
    +facet_grid( 'class~.' )
    +scale_fill_hue(s = 0.90,l= 0.65,h=0.0417,color_space="husl")
    +theme_light()
    +theme(aspect_ratio =0.25,
    dpi=100,
    figure_size=(3,4)
    ))
print(density_plot)
```

```python
jitter_plot = (ggplot(df, aes(x="class", y="value", fill="class"))
               + geom_jitter(width=0.3, size=3, stroke=0.1, show_legend=False)
               + scale_fill_hue(s=0.90, l=0.65, h=0.0417, color_space='husl')
               + theme_matplotlib()
               + theme(aspect_ratio=1.05,
                       dpi=100,
                       figure_size=(4, 4)
                       ))
print(jitter_plot)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_palette("husl")#设定绘图的颜色主题
fig = plt.figure(figsize=(6,4),dpi=100)
sns.swarmplot(x="class",y="value",hue="class",data=df,edgecolor='k',linewidth=0.2)
plt.legend().set_visible(False)
```

```python
dot_plot = (ggplot(df, aes(x='class', y="value", fill="class"))
               + geom_dotplot(binaxis="y", stackdir='center',binwidth=0.15, show_legend=False)
               + scale_fill_hue(s=0.90, l=0.65, h=0.0417, color_space='husl')
               + theme_matplotlib()
               + theme(aspect_ratio=1.05, dpi=100, figure_size=(4, 4)))
print(dot_plot)
```

```python
jitter_plot = (ggplot(df, aes(x='class', y="value", fill="class"))
               + geom_jitter(width=0.3, size=3, stroke=0.1, show_legend=False)
               + stat_summary(fun_data="mean_sdl", fun_args={
                              'mult': 1}, geom="pointrange", fill="w", color="black", size=1, show_legend=False)
               + scale_fill_hue(s=0.90, l=0.65, h=0.0417, color_space='husl')
               + theme_matplotlib()
               + theme(
    aspect_ratio=1.05,
    dpi=100,
    figure_size=(4, 4))
)
print(jitter_plot)
```
