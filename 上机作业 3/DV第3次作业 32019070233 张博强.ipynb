```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = x ** 2
plt.figure(num=1, figsize=(10, 4))
plt.plot(x, y1)
plt.plot(x, y2, color='red', linewidth=3, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-2, -1, 0, 2, 3], [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
plt.show()
```

```python
n = np.linspace(-5, 4, 30)
m1 = 3 * n + 2
m2 = n ** 2

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

l1, = plt.plot(n, m1, label='up')
l2, = plt.plot(n, m2, label='down', color='blue', linewidth=1, linestyle='-')
plt.xlim((-5, 4))
plt.ylim((-5, 15))
plt.yticks([10, -3], [r'$love$', r'hate'])
plt.xlabel("XXX")
plt.ylabel("YYY")
plt.show()
```

```python
plt.figure(num=5, figsize=(8, 5))
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
plt.plot(x, y1)
x0 = 1
y0 = 2 * x0 + 1
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
plt.scatter(x0, y0, s=50, color='b')
plt.plot([x0, x0], [y0, 0], 'k--', linewidth=2.5)
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data',
             xytext=(+30, -30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))
plt.text(-3.7, 3, r'$This\ is\ the\ some\ text.\ \mu\sigma_i\ \alpha_t$', fontdict={'size': 16, 'color': 'r'})
```

```python
t = np.linspace(0, 5, 300)
s = np.cos(2*np.pi*t)
x0 = 1
y0 = np.cos(2*np.pi*1)
plt.ylim([-2,2])
plt.annotate(r'$local\ max$' % y0, xy=(x0, y0), xycoords='data',
             xytext=(+60, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->',linewidth=5))
plt.plot(t,s)
```

```python
x = np.linspace(0, 1, 500)
y = np.sin(3 * np.pi * x) * np.exp(-4 * x)
plt.plot(x, y)
plt.fill_between(x, 0, y, facecolor='green', alpha=0.3)
plt.show()
```

```python
x = np.linspace(0, 1, 500)
y1 = np.sin(3 * np.pi * x) * np.exp(-4 * x)
y2 = y1 + 0.2
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r')
plt.fill_between(x, y1, y2, facecolor='green', alpha=0.3)
plt.show()
```

```python
fig, ax = plt.subplots()


def f(t):
    return np.cos(2 * np.pi * t)


x1 = np.arange(0.0, 4.0, 0.5)
x2 = np.arange(0.0, 4.0, 0.01)
plt.figure(1)
plt.subplot(2, 2, 1)
plt.xticks([0, 2, 4])
plt.plot(x1, f(x1), 'bo', x2, f(x2), 'k')
plt.subplot(2, 2, 2)
plt.plot(np.cos(2 * np.pi * x2), 'r--')
plt.xticks([0, 200, 400])
plt.show()
```
