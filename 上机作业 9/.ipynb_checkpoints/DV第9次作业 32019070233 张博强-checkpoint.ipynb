```python
import pandas as pd
import numpy as np
from plotnine import *
df = pd.read_csv('Hist_Density_Data.csv')
df_al = df.loc[df['Location']=='Adelaide']
base_hist=(ggplot(df_al, aes(x='MXSPD', fill='Location'))
    +geom_histogram(binwidth = 1, alpha=0.55, colour="black",size=0.25)
    +scale_fill_hue(s = 0.90, l = 0.65, h=0.0417,color_space='husl')
    +theme(
        text=element_text(size=13,color="black"),
        plot_title=element_text(size=15,family="myfont",face="bold.italic",hjust=.5,color="black"),
        legend_position=(0.7,0.8),
        legend_background = element_blank(),
        aspect_ratio =1.15,
        figure_size=(5,5)
    ))
print(base_hist)
```

```python
base_density = (ggplot(df_al, aes(x='MXSPD', fill='Location'))
                + geom_density(bw=1, alpha=0.55, colour="black",
                               size=0.25, kernel="gaussian")
                + scale_fill_hue(s=0.90, l=0.65, h=0.0417, color_space='husl')
                + theme(
    text=element_text(size=13, color="black"),
    plot_title=element_text(size=15, family="myfont",
                            face="bold.italic", hjust=.5, color="black"),
    legend_position=(0.7, 0.75),
    legend_background=element_blank(),
    aspect_ratio=1.15,
    figure_size=(5, 5)
))
print(base_density)
```

```python
import pandas as pd
import joypy
import matplotlib.pyplot as plt
from matplotlib import cm
from pandas.api.types import CategoricalDtype
df = pd.read_csv("lincoln_weather.csv")
df_type=CategoricalDtype(categories=['January', 'February', 'March', 'April', 'May', 'June', 'July',
'August', 'September', 'October', 'November','December'], ordered=True)
df['Month'] = df['Month'].astype(df_type)
fig, axes = joypy.joyplot(df, column=["Mean.Temperature..F."], by="Month", ylim='own',
colormap=cm.Spectral_r, alpha= 0.9,figsize=(6,5))
plt.xlabel("Mean Temperature",{'size': 15 })
plt.ylabel("Month",{'size': 15 })
```
